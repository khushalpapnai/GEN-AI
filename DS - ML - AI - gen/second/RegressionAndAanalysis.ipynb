```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pyl
from sklearn.linear_model import LinearRegression
```

```python
data =  pd.read_csv("lsd_math_score_data.csv")
```

```python
data
```

```python
time = data[["Time_Delay_in_Minutes"]]
lsd = data[["LSD_ppm"]]
score = data[["Avg_Math_Test_Score"]]
```

```python
%matplotlib inline
pyl1 = pyl
pyl1.plot(lsd,time,color = "red")
pyl1.plot(lsd,score,color = "yellow")
pyl1.text(x=2.7,y=-150,s="wagner et al. (1968)",fontsize=12)
pyl1.title("amount of drug increase inside body that affect in score")
pyl1.xlabel("drug tissue",fontsize = 12)
pyl1.xticks(fontsize = 10)
pyl1.ylabel("time in minutes and scores",fontsize = 12)
pyl1.yticks(fontsize = 10)
pyl1.ylim(0,500)
pyl1.style.use('dark_background')
pyl1.show()
```

```python
reg = LinearRegression()
reg.fit(lsd,score)
print("theta :", reg.coef_[0][0])
print("Intercept :",reg.intercept_[0])
print("R-square :",reg.score(lsd,score))
```

```python
%matplotlib inline
pyl2 = pyl
pyl2.scatter(lsd,score,color = "blue",s = 50,alpha= 0.7)
pyl2.plot(lsd,reg.predict((lsd)),color = "m")
pyl2.text(x=2.7,y=15,s="wagner et al. (1968)",fontsize=12,color = "y")
pyl2.title("Scores affected by drug ",y = 1.1 )
pyl2.xlabel("Drug Tissue",fontsize = 12,color = "red")
pyl2.xticks(fontsize = 10)
pyl2.ylabel("Avg Scores",fontsize = 12,color = "red")
pyl2.yticks(fontsize = 10)
pyl2.style.use('dark_background')
pyl2.rcParams['grid.linestyle'] = '--'
pyl2.rcParams['grid.alpha'] = 1
pyl2.show()
```
