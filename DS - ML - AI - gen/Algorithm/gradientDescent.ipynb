### Gradient Descent

```python
 # Gradient Descent
new_x = 4
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [der(new_x)]


for n in range(500):
    pre_x = new_x
    gradient = der(pre_x)
    new_x = pre_x - step_multiplier * gradient

    x_list.append(new_x)
    slope_list.append(der(new_x))

    step_size = abs(new_x - pre_x)
    if step_size < precision:
        print(step_size,n)
        break

print("local minimum occurs at:", new_x)
print("slope or der(x) value at this point is:", der(new_x))
print("f(x) value of cost at this point is:", f(new_x))

```

### Gradient descent as a python funtion


```python
# Gradient Descent

def GD(derivative_func,initial,multiplier=0.02,precision=0.001,mat_iter = 10):
    new_x = initial

    x_list = [new_x]
    slope_list = [derivative_func(new_x)]


    for n in range(mat_iter):
        pre_x = new_x
        gradient = derivative_func(pre_x)
        new_x = pre_x - multiplier * gradient

        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        step_size = abs(new_x - pre_x)
        if step_size < precision:
            break
    return x_list,slope_list,new_x
```

```python
 # while Gradient Descent
def while_gd(derivative_func,initial,multiplier=0.02,precision=0.001,mat_iter = 71):
    new_x = initial

    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    n = 0
    while n <=  mat_iter:
        pre_x = new_x
        gradient = derivative_func(pre_x)
        new_x = pre_x - multiplier * gradient

        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        step_size = abs(new_x - pre_x)
        if step_size < precision:
            break
        n += 1
    return x_list,slope_list,new_x
```

### Batch Gradient Descent with SymPy

```python
# setup

multiplier = 0.1
max_iter = 500
prams = np.array([1.8,1.0]) # initial guess

for n in range(max_iter):
    gradient_x = diff(f3d(a,b),a).evalf(subs={a:prams[0],b:prams[1]})
    gradient_y = diff(f3d(a,b),b).evalf(subs={a:prams[0],b:prams[1]})
    gradients = np.array([gradient_x,gradient_y])
    prams = prams - multiplier * gradients

# Results
print("Values in gradient array", gradients)
print("Minimum occurs at X value of: ", prams[0])
print("Minimum occurs at Y value of: ", prams[1])
print("The cost is: ", f3d(prams[0],prams[1]))
```

### Partial Derivatives & Symbolic Computation
$$ \frac{\partial f}{\partial x} = \frac{2x \ln(3) \cdot 3^{-x^2-y^2}} {\left( 3^{-x^2-y^2}+1 \right)^2}  $$
$$ \frac{\partial f}{\partial y} = \frac{2y \ln(3) \cdot 3^{-x^2-y^2}} {\left( 3^{-x^2-y^2}+1 \right)^2}  $$

```python
def pdx(x,y):
    r = 3**(-x**2-y**2)
    return 2*x*log(3)*r / (r+1)**2

def pdy(x,y):
    r = 3**(-x**2 - y**2)
    return 2*y*log(3)*r / (r+1)**2
```

```python
pdx(1.8,1.0)
```

```python
# setup

multiplier = 0.1
max_iter = 500
prams = np.array([1.8,1.0]) # initial guess

for n in range(max_iter):
    gradient_x = pdx(prams[0],prams[1])
    gradient_y = pdy(prams[0],prams[1])
    gradients = np.array([gradient_x,gradient_y])
    prams = prams - multiplier * gradients

# Results
print("Values in gradient array", gradients)
print("Minimum occurs at X value of: ", prams[0])
print("Minimum occurs at Y value of: ", prams[1])
print("The cost is: ", f3d(prams[0],prams[1]))
```

#### Mean squared error: A cost function for regressio problem


$$ RSS = \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$
$$ MSE = \frac {1}{n} \cdot \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$
$$ MSE = \frac {1}{n} \cdot \sum_{i=1}^{n} \big( y - \hat{y} \big)^2 $$

### function of MSE ( mean squared method )


```python
def MSE(y_5,y_hat):
    # return sum((y_5-y_hat)**2) / y_5.size
    return np.average((y_5-y_hat)**2,axis=0)
```

```python
print("Manually calculation",MSE(y_5,y_hat)[0])
print("MSE regression using manual calc is ",mean_squared_error(y_5,y_hat))
print("MSE regression is ",mean_squared_error(y_5,model.predict(x_5)))
```

### Partial Derivatives of MSE w.r.t. $ \theta_0$ and $ \theta_1 $

$$ \frac{\partial MSE} {\partial \theta_0} = - \frac {2} {n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) $$
$$ \frac{\partial MSE} {\partial \theta_1} = - \frac {2} {n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)}  \big) \big( x^{(i)} \big) $$


### MSE & Gradient Descent

```python
# x values, y values, array of theta parameter { theta0 at index 0 and theta theta1 at index }

def msegd(x,y,thetas):
    n  = y.size
    slope1 = (-2/n) * sum(y-thetas[0]-thetas[1]*x)
    slope2 = (-2/n) * sum((y-thetas[0]-thetas[1]*x)*x)
#    return np.array([slope1[0],slope2[0]])
    return np.append(arr = slope1, values = slope2)
#     return np.concatenate((slope1,slope2))
```

```python
multiplier = 0.01
thetas = np.array([2.9,2.9])

plot_vals = thetas.reshape(1, 2)
mse_vals = MSE(y_5,thetas[0]+thetas[1] * x_5)

for i in range (1000):
    thetas = thetas - multiplier * msegd(x_5,y_5,thetas)
    plot_vals = np.concatenate((plot_vals,thetas.reshape(1,2)),axis = 0)
    mse_vals = np.append(arr=mse_vals, values = MSE(y_5,thetas[0]+thetas[1] * x_5))
# results
print("min occurs at theta 0:", thetas[0])
print("min occurs at theta 1:", thetas[1])
print("mse is:",MSE(y_5,thetas[0]+thetas[1]*x_5))
print("mse is:",mse_vals)
print("plot is:",plot_vals)
```
