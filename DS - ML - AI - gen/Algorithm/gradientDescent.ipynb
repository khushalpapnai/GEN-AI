## Notebook import and packages

```python
import matplotlib.pyplot as pyl
import numpy as np
%matplotlib inline
```

### Example 1 - A Simple cost function

$$f(x) = x^2 + x + 1$$

```python
# Example 1
def f(x):
    return x**2 + x + 1
```

```python
# Make Data
data1 = np.linspace(start=-2,stop=2,num = 500)
```

```python
# Plot

pyl.title("Gradient Descent Algorithm")
pyl.xlabel("x",fontsize = 15, fontweight="bold", color = "red", fontstyle = "italic" )
pyl.ylabel("f(x)",fontsize = 15, fontweight="bold", color = "green", fontstyle="italic")
pyl.style.use('dark_background')
pyl.rcParams['grid.linestyle'] = '--'
pyl.rcParams['grid.alpha'] = 0.7
pyl.rcParams["grid.color"] = "y"

pyl.plot(data1,f(data1))
pyl.grid(True)
pyl.show()
```

### Slope & Derivatives


```python
def der(x):
    return 2*x + 1
```

```python
# Plot Function and Derivative side by side

pyl.figure(figsize=[10,5])
pyl.subplot(1,2,1)

# add detail
pyl.xlim(-2,2)
pyl.ylim(0,8)
pyl.title("Cost Function")
pyl.xlabel("x",fontsize = 15, fontweight="bold", color = "red", fontstyle = "italic" )
pyl.ylabel("f(x)",fontsize = 15, fontweight="bold", color = "green", fontstyle="italic")
pyl.style.use('dark_background')
pyl.style.use('dark_background')
pyl.rcParams['grid.linestyle'] = '--'
pyl.rcParams['grid.alpha'] = 0.7
pyl.rcParams["grid.color"] = "y"
pyl.grid(True)
pyl.plot(data1,f(data1),color = "blue",linewidth = 5)

pyl.subplot(1,2,2)
pyl.xlim(-2,2)
pyl.ylim(-3,5)
pyl.title("Slope of the Cost Function")
pyl.xlabel("x",fontsize = 15, fontweight="bold", color = "red", fontstyle = "italic" )
pyl.ylabel("der(x)",fontsize = 15, fontweight="bold", color = "green", fontstyle="italic")
pyl.style.use('dark_background')
pyl.style.use('dark_background')
pyl.rcParams['grid.linestyle'] = '--'
pyl.rcParams['grid.alpha'] = 0.7
pyl.rcParams["grid.color"] = "y"
pyl.plot(data1,der(data1),color = "skyblue",linewidth = 5)

pyl.grid(True)
pyl.show()
```
