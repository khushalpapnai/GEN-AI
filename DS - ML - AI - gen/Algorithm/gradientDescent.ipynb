### Slope & Derivatives
$$ der(x) = 2*x + 1 $$


```python
def der(x):
    return 2*x + 1
```

### Gradient Descent

```python
 # Gradient Descent
new_x = 4
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [der(new_x)]


for n in range(500):
    pre_x = new_x
    gradient = der(pre_x)
    new_x = pre_x - step_multiplier * gradient

    x_list.append(new_x)
    slope_list.append(der(new_x))

    step_size = abs(new_x - pre_x)
    if step_size < precision:
        print(step_size,n)
        break

print("local minimum occurs at:", new_x)
print("slope or der(x) value at this point is:", der(new_x))
print("f(x) value of cost at this point is:", f(new_x))

```

### Gradient descent as a python funtion


```python
# Gradient Descent

def GD(derivative_func,initial,multiplier=0.02,precision=0.001,mat_iter = 10):
    new_x = initial

    x_list = [new_x]
    slope_list = [derivative_func(new_x)]


    for n in range(mat_iter):
        pre_x = new_x
        gradient = derivative_func(pre_x)
        new_x = pre_x - multiplier * gradient

        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        step_size = abs(new_x - pre_x)
        if step_size < precision:
            break
    return x_list,slope_list,new_x
```

```python
 # while Gradient Descent
def while_gd(derivative_func,initial,multiplier=0.02,precision=0.001,mat_iter = 71):
    new_x = initial

    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    n = 0
    while n <=  mat_iter:
        pre_x = new_x
        gradient = derivative_func(pre_x)
        new_x = pre_x - multiplier * gradient

        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        step_size = abs(new_x - pre_x)
        if step_size < precision:
            break
        n += 1
    return x_list,slope_list,new_x
```

### Example 4 - Data Visualization with 3D Charts
### minimise
$$ F3D(x,y) = \frac{1}{3^{-x^2-y^2}+1} $$

```python
# Function of 3D Viz
def f3d(x,y):
    r = 3**(-x**2-y**2)
    return 1 / (r + 1)
```

```python
x_4 = np.linspace(-2,2,200)
y_4 = np.linspace(-2,2,200)

x_4,y_4 = np.meshgrid(x_4,y_4)
print("shape of x array",x_4.shape)
```

```python
# Generating 3D Plot

fig = pyl.figure(figsize=[16,12])
ax = fig.add_subplot(111,projection='3d')

ax.set_xlabel("X",color = "r", fontsize = 15)
ax.set_ylabel("Y",color = "r", fontsize = 15)
ax.set_zlabel("f3d(x,y) - cost",color = "r", fontsize = 15)

ax.plot_surface(x_4,y_4,f3d(x_4,y_4),cmap='BuPu',alpha = 0.5)
```

### partial Derivatives & Symbolic Computation

```python
a,b = symbols('x,y')
print("Our cost function f3d(x,y) is: ",f3d(a,b))
print("partial derivative wrt x is: ",diff(f3d(a,b),a))
print("value of f3d(x,y) at x = 1.8 , y = 1.0 is: ",
      f3d(a,b).evalf(subs={a:1.8,b:1}) )# Python dictionary
print("value of partial derivative wrt x: ",diff(f3d(a,b),a).evalf(subs={a:1.8,b:1}))
```

### Batch Gradient Descent with SymPy

```python
# setup

multiplier = 0.1
max_iter = 500
prams = np.array([1.8,1.0]) # initial guess

for n in range(max_iter):
    gradient_x = diff(f3d(a,b),a).evalf(subs={a:prams[0],b:prams[1]})
    gradient_y = diff(f3d(a,b),b).evalf(subs={a:prams[0],b:prams[1]})
    gradients = np.array([gradient_x,gradient_y])
    prams = prams - multiplier * gradients

# Results
print("Values in gradient array", gradients)
print("Minimum occurs at X value of: ", prams[0])
print("Minimum occurs at Y value of: ", prams[1])
print("The cost is: ", f3d(prams[0],prams[1]))
```
