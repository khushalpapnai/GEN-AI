### Example 1 - A Simple cost function

$$f(x) = x^2 + x + 1$$

```python
# Example 1
def f(x):
    return x**2 + x + 1
```

```python
# Make Data
data1 = np.linspace(start=-2,stop=2,num = 500)
```

```python
# Plot

pyl.title("Gradient Descent Algorithm")
pyl.xlabel("x",fontsize = 15, fontweight="bold", color = "red", fontstyle = "italic" )
pyl.ylabel("f(x)",fontsize = 15, fontweight="bold", color = "green", fontstyle="italic")
pyl.style.use('dark_background')
pyl.rcParams['grid.linestyle'] = '--'
pyl.rcParams['grid.alpha'] = 0.7
pyl.rcParams["grid.color"] = "y"

pyl.plot(data1,f(data1))
pyl.grid(True)
pyl.show()
```

### Slope & Derivatives
$$ der(x) = 2*x + 1 $$


```python
def der(x):
    return 2*x + 1
```

```python
# Plot Function and Derivative side by side

pyl.figure(figsize=[10,5])
pyl.subplot(1,2,1)

# add detail
pyl.xlim(-2,2)
pyl.ylim(0,8)
pyl.title("Cost Function")
pyl.xlabel("x",fontsize = 15, fontweight="bold", color = "red", fontstyle = "italic" )
pyl.ylabel("f(x)",fontsize = 15, fontweight="bold", color = "green", fontstyle="italic")
pyl.style.use('dark_background')
pyl.style.use('dark_background')
pyl.rcParams['grid.linestyle'] = '--'
pyl.rcParams['grid.alpha'] = 0.7
pyl.rcParams["grid.color"] = "y"
pyl.grid(True)
pyl.plot(data1,f(data1),color = "blue",linewidth = 5)

pyl.subplot(1,2,2)
pyl.xlim(-2,2)
pyl.ylim(-3,5)
pyl.title("Slope of the Cost Function")
pyl.xlabel("x",fontsize = 15, fontweight="bold", color = "red", fontstyle = "italic" )
pyl.ylabel("der(x)",fontsize = 15, fontweight="bold", color = "green", fontstyle="italic")
pyl.style.use('dark_background')
pyl.style.use('dark_background')
pyl.rcParams['grid.linestyle'] = '--'
pyl.rcParams['grid.alpha'] = 0.7
pyl.rcParams["grid.color"] = "y"
pyl.plot(data1,der(data1),color = "skyblue",linewidth = 5)

pyl.grid(True)
pyl.show()
```

### Gradient Descent

```python
 # Gradient Descent
new_x = 4
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [der(new_x)]


for n in range(500):
    pre_x = new_x
    gradient = der(pre_x)
    new_x = pre_x - step_multiplier * gradient

    x_list.append(new_x)
    slope_list.append(der(new_x))

    step_size = abs(new_x - pre_x)
    if step_size < precision:
        print(step_size,n)
        break

print("local minimum occurs at:", new_x)
print("slope or der(x) value at this point is:", der(new_x))
print("f(x) value of cost at this point is:", f(new_x))

```

```python
# Superimpose gradient decent calculation on plot

pyl.figure(figsize=[10,5])


# Cost Function
pyl.subplot(1,2,1)
pyl.xlim(-2,2)
pyl.ylim(0,8)
pyl.title("Cost Function")
pyl.xlabel("x",fontsize = 15, fontweight="bold", color = "red", fontstyle = "italic" )
pyl.ylabel("f(x)",fontsize = 15, fontweight="bold", color = "green", fontstyle="italic")
pyl.style.use('dark_background')
pyl.style.use('dark_background')
pyl.rcParams['grid.linestyle'] = '--'
pyl.rcParams['grid.alpha'] = 0.7
pyl.rcParams["grid.color"] = "y"
pyl.grid(True)
pyl.plot(data1,f(data1),color = "blue",linewidth = 5,alpha = 0.5)
values = np.array(x_list)
pyl.scatter(values,f(values),color = "m",s = 100,alpha = 0.7)


# Derivative
pyl.subplot(1,2,2)
pyl.xlim(-2,2)
pyl.ylim(-3,5)
pyl.title("Slope of the Cost Function")
pyl.xlabel("x",fontsize = 15, fontweight="bold", color = "red", fontstyle = "italic" )
pyl.ylabel("der(x)",fontsize = 15, fontweight="bold", color = "green", fontstyle="italic")
pyl.style.use('dark_background')
pyl.style.use('dark_background')
pyl.rcParams['grid.linestyle'] = '--'
pyl.rcParams['grid.alpha'] = 0.7
pyl.rcParams["grid.color"] = "y"
pyl.plot(data1,der(data1),color = "skyblue",linewidth = 5,alpha = 0.5)
pyl.scatter(x_list,slope_list, color = "m",s = 100,alpha=0.7)

pyl.grid(True)
pyl.show()
```

### Example 2 Multiple Minima vs Initial Guess & Advanced Function
$$ G(x) = x^4 - 4*x^2 + 5 $$

```python
def g(x):
    return x**4 - 4 * x**2 + 5
```

```python
def dg(x):
    return 4*x**3 - 8 * x
```

```python
# Make some data
data2 = np.linspace(-2,2,num = 1000)
```

```python
pyl.figure(figsize=[18,5])
pyl.subplot(1,3,1)
pyl.xlim(-2,2)
pyl.ylim(0.5,5)
pyl.title("Gradient Descent Algorithm")
pyl.xlabel("x",color = "r", fontsize = 15)
pyl.ylabel("G(x",color = "g", fontsize = 15)

pyl.style.use("dark_background")
pyl.rcParams['grid.linestyle'] = '--'
pyl.rcParams["grid.color"] = "y"

pyl.grid(True)
pyl.plot(data2,g(data2),color = "b", linewidth = 5)



pyl.subplot(1,3,2)
pyl.xlim(-2,2)
pyl.ylim(-6,8)
pyl.title("Gradient Descent Algorithm")
pyl.xlabel("x",color = "r", fontsize = 15)
pyl.ylabel("DG(x",color = "g", fontsize = 15)

pyl.style.use("dark_background")
pyl.rcParams["grid.linestyle"] = "--"
pyl.rcParams["grid.color"] = "y"

pyl.grid(True)
pyl.plot(data2,dg(data2),color = "skyblue", linewidth = 5)

pyl.show()
```

### Gradient descent as a python funtion


```python
# Gradient Descent

def GD(derivative_func,initial,multiplier=0.02,precision=0.001,mat_iter = 10):
    new_x = initial

    x_list = [new_x]
    slope_list = [derivative_func(new_x)]


    for n in range(mat_iter):
        pre_x = new_x
        gradient = derivative_func(pre_x)
        new_x = pre_x - multiplier * gradient

        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        step_size = abs(new_x - pre_x)
        if step_size < precision:
            break
    return x_list,slope_list,new_x
```

```python
list_x,der_x,local_min = GD(derivative_func=dg,initial=0.5,multiplier=0.02,precision=0.001)
print("local min occurs at",len(list_x),"\nNumber of steps",local_min)
```

```python
list_x,der_x,local_min = GD(derivative_func=dg,initial=0.1)
pyl.figure(figsize=[18,5])
pyl.subplot(1,3,1)
pyl.xlim(-2,2)
pyl.ylim(0.5,5)
pyl.title("Gradient Descent Algorithm")
pyl.xlabel("x",color = "r", fontsize = 15)
pyl.ylabel("G(x",color = "g", fontsize = 15)

pyl.style.use("dark_background")
pyl.rcParams['grid.linestyle'] = '--'
pyl.rcParams["grid.color"] = "y"

pyl.grid(True)
pyl.plot(data2,g(data2),color = "b", linewidth = 5, alpha = 0.6)
values = np.array(list_x)
pyl.scatter(list_x,g(values),color = "m",s = 100,alpha=0.7)


pyl.subplot(1,3,2)
pyl.xlim(-2,2)
pyl.ylim(-6,8)
pyl.title("Gradient Descent Algorithm")
pyl.xlabel("x",color = "r", fontsize = 15)
pyl.ylabel("DG(x",color = "g", fontsize = 15)

pyl.style.use("dark_background")
pyl.rcParams["grid.linestyle"] = "--"
pyl.rcParams["grid.color"] = "y"

pyl.grid(True)
pyl.plot(data2,dg(data2),color = "skyblue", linewidth = 5, alpha = 0.6)
pyl.scatter(list_x,der_x,color = "m",s = 100,alpha=0.7)

pyl.show()
```

### Example 3 Divergence, overflow and python tuples
$$ H(x) = x^5 - 2x^4 + 2 $$


```python
# make data
data3 = np.linspace(-2.5,2.5,num = 1000)
```

```python
def h(x):
    return x**5 - 2*x**4 + 2
```

```python
def dh(x):
    return 5*x**4 - 8*x**3
```

```python
hx_list,dhx_list,h_x = GD(derivative_func=dh,initial=-0.2, mat_iter=71)
pyl.figure(figsize=[18,5])
pyl.subplot(1,3,1)
pyl.xlim(-1.5,2.5)
pyl.ylim(-1,5)
pyl.title("Gradient Descent Algorithm")
pyl.xlabel("x",color = "r", fontsize = 15)
pyl.ylabel("G(x",color = "g", fontsize = 15)

pyl.style.use("dark_background")
pyl.rcParams['grid.linestyle'] = '--'
pyl.rcParams["grid.color"] = "y"

pyl.grid(True)
pyl.plot(data3,h(data3),color = "b", linewidth = 5, alpha = 0.6)
pyl.scatter(hx_list,h(np.array(hx_list)),color = "m",s = 100,alpha=0.7)



pyl.subplot(1,3,2)
pyl.xlim(-1.5,2.5)
pyl.ylim(-4,5)
pyl.title("Gradient Descent Algorithm")
pyl.xlabel("x",color = "r", fontsize = 15)
pyl.ylabel("DG(x",color = "g", fontsize = 15)

pyl.style.use("dark_background")
pyl.rcParams["grid.linestyle"] = "--"
pyl.rcParams["grid.color"] = "y"

pyl.grid(True)
pyl.plot(data3,dh(data3),color = "skyblue", linewidth = 5)
pyl.scatter(hx_list,dhx_list,color = "m",s = 100,alpha=0.7)

pyl.show()
print("local min occurs at:", h_x)
print("cost at this minimum is:", h(h_x))
print("Number of steps:",len(hx_list))
```
